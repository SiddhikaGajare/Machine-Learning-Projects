```python
import pandas as pd
```

```python
sns = pd.read_csv('snsdata.csv')
sns.head()
```

```python
sns.gender = sns.gender.fillna('F')
sns.age = sns.age.fillna(sns.age.mean())
sns.gender = sns.gender.replace({'F':1, 'M':0})
```

```python
#sns.isnull().sum()
```

```python
#sns.info()
```

```python
sns.head()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler, normalize
```

### While building the DBSCAN model, you need to pass 2 hyper parameter
1. Min Distance
2. Min number of points

As we need to pass the data, we will first standardise and normalize it

```python
scalar = StandardScaler()
sns1 = scalar.fit_transform(sns)
```

```python
normalize_sns = normalize(sns1)
```

```python
dbscan_sns = DBSCAN(min_samples=4, eps=0.6)
```

```python
dbscan_sns.fit(normalize_sns)   # Model created
```

```python
#list(dbscan_sns.labels_)
```

```python
set(dbscan_sns.labels_)
```

```python
abcd = pd.Series(dbscan_sns.labels_)
abcd.value_counts() # -1 indicates noise points
```
